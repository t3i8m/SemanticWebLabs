```python
from rdflib import Graph, Literal, Namespace
```

```python
g = Graph()
g.parse("5_ken3140_webshop.ttl", format="ttl")
```

```python
qres = """
SELECT DISTINCT ?s ?p ?o
WHERE {

  ?s ?p ?o .

} LIMIT 100
"""

for row in g.query(qres):
    print(f"{row.s} {row.p} {row.o}")
```

```python
# For a given item (select a random item from your RDF Graph), provide all its
# categories and subcategories, and its brand

query_a = """
PREFIX schema: <https://schema.org/> 
PREFIX vinted: <http://semanticweb2025.group5/vinted/>

SELECT ?item_name ?primary_category ?secondary_category ?item_brand
WHERE {

  ?item schema:name ?item_name;
        vinted:primaryCategory ?primary_category;
        vinted:secondaryCategory ?secondary_category;
        schema:brand ?item_brand.


} LIMIT 1
"""
for row in g.query(query_a):
    print(f"Name: {row.item_name}\nCategory: {row.primary_category}\nSubCategory: {row.secondary_category}\nBrand: {row.item_brand}")
```

```python
# Provide items from different subcategories that have the same brand.
query_b = """
PREFIX schema: <https://schema.org/> 
PREFIX vinted: <http://semanticweb2025.group5/vinted/>

SELECT ?brand_name ?item_name_1 ?item_subcategory_1 ?item_name_2 ?item_subcategory_2
WHERE {

    ?brand schema:name ?brand_name.

    ?item schema:name ?item_name_1;
        vinted:primaryCategory ?item_category_1;
        vinted:secondaryCategory ?item_subcategory_1;
        schema:brand ?brand.
        
    ?item2 schema:name ?item_name_2;
        vinted:primaryCategory ?item_category_2;
        vinted:secondaryCategory ?item_subcategory_2;
        schema:brand ?brand.


    FILTER(?item_subcategory_2!=?item_subcategory_1)

} 
"""

for row in g.query(query_b):
    print("\n")
    print(f"Brand: {row.brand_name}\nItem_1: {row.item_name_1}\nSubCategory_Item_1: {row.item_subcategory_1}\nItem_2: {row.item_name_2}\nSubCategory_Item_2: {row.item_subcategory_2}")
```

```python
# Group products by brand and show the average price or rating for each brand.
query_c = """
PREFIX schema: <https://schema.org/> 
PREFIX vinted: <http://semanticweb2025.group5/vinted/>

SELECT ?brand_name (AVG(?item_price) AS ?avg_price)
WHERE {

    ?brand schema:name ?brand_name.

    ?item schema:brand ?brand;
        schema:price ?item_price.



} GROUP BY ?brand_name
LIMIT 10
"""

for row in g.query(query_c):
    print(f"{row.brand_name} {row.avg_price}")
```

```python
# Sort products in one category according to average brand price or rating.
query_d = """
PREFIX schema: <https://schema.org/> 
PREFIX vinted: <http://semanticweb2025.group5/vinted/>

SELECT ?brand_name (AVG(?item_price) AS ?avg_price)
WHERE {
  ?item vinted:primaryCategory vinted:Men ;
        schema:name ?item_name ;
        schema:brand ?brand ;
        schema:price ?item_price .
  ?brand schema:name ?brand_name .
}
GROUP BY  ?brand_name
ORDER BY ?avg_price
LIMIT 10
"""

for row in g.query(query_d):
    print(f"\nBrand:{row.brand_name}\nAvg_price:{row.avg_price}")
```

```python
# Brands in one category (Women) sorted by how many items they have + the brand's average price

# This query shows which brands have the largest Men assortment and their average price, enabling smarter ranking, 
# filters, and brand curation. It also informs pricing and promotions by revealing typical price points per brand, 
# improving user expectations and conversion.

query_g = """
PREFIX schema: <https://schema.org/> 
PREFIX vinted: <http://semanticweb2025.group5/vinted/>
PREFIX xsd:    <http://www.w3.org/2001/XMLSchema#>

SELECT ?brand_name (COUNT(?item) AS ?num_items) (AVG(xsd:decimal(?item_price)) AS ?avg_price)
WHERE {
  ?item vinted:primaryCategory vinted:Women ;
        schema:brand ?brand ;
        schema:price ?item_price .
  ?brand schema:name ?brand_name .
}
GROUP BY ?brand_name
ORDER BY DESC(?num_items) ?avg_price
LIMIT 10
"""

for row in g.query(query_g):
    print(f"{row.brand_name} {row.num_items} {row.avg_price}")
```

```python

from SPARQLWrapper import SPARQLWrapper, JSON

sparql_endpoint = "https://query.wikidata.org/sparql"

query_e = """
PREFIX wdt: <http://www.wikidata.org/prop/direct/>
PREFIX wd: <http://www.wikidata.org/entity/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX bd: <http://www.bigdata.com/rdf#>
PREFIX wikibase: <http://wikiba.se/ontology#>

SELECT ?brandLabel ?inceptionDate ?founderLabel ?headquartersLabel (GROUP_CONCAT(DISTINCT ?industryLabel; SEPARATOR=", ") AS ?industries) ?officialWebsite
WHERE {
  BIND(wd:Q3895 AS ?brand)

  OPTIONAL { ?brand wdt:P571 ?inceptionDate. }
  OPTIONAL { ?brand wdt:P112 ?founder. }
  OPTIONAL { ?brand wdt:P159 ?headquarters. }
  OPTIONAL { ?brand wdt:P856 ?officialWebsite. }
  OPTIONAL { ?brand wdt:P452 ?industry. }

  SERVICE wikibase:label {
    bd:serviceParam wikibase:language "en".
    ?brand rdfs:label ?brandLabel.
    ?founder rdfs:label ?founderLabel.
    ?headquarters rdfs:label ?headquartersLabel.
    ?industry rdfs:label ?industryLabel. 
  }
}
GROUP BY ?brandLabel ?inceptionDate ?founderLabel ?headquartersLabel ?officialWebsite
LIMIT 1
"""

sparql = SPARQLWrapper(sparql_endpoint)
sparql.setQuery(query_e)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
brand_data = results["results"]["bindings"][0]

name = brand_data.get("brandLabel", {}).get("value")
inception = brand_data.get("inceptionDate", {}).get("value")
founder = brand_data.get("founderLabel", {}).get("value")
hq = brand_data.get("headquartersLabel", {}).get("value")
industries = brand_data.get("industries", {}).get("value")
website = brand_data.get("officialWebsite", {}).get("value")

print(f"Brand: {name}")
print(f"Inception: {inception}")
print(f"Founder: {founder}")
print(f"Headquarters: {hq}")
print(f"Industries: {industries}")
print(f"Website: {website}")

```
