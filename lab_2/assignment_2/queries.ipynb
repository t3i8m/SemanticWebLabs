```python
from rdflib import Graph, Literal, Namespace
```

```python
g = Graph()
g.parse("5_ken3140_webshop.ttl", format="ttl")
```

```python
qres = """
SELECT DISTINCT ?s ?p ?o
WHERE {

  ?s ?p ?o .

} LIMIT 100
"""

for row in g.query(qres):
    print(f"{row.s} {row.p} {row.o}")
```

```python
# For a given item (select a random item from your RDF Graph), provide all its
# categories and subcategories, and its brand

query_a = """
PREFIX schema: <https://schema.org/> 
PREFIX vinted: <http://semanticweb2025.group5/vinted/>

SELECT ?item_name ?primary_category ?secondary_category ?item_brand
WHERE {

  ?item schema:name ?item_name;
        vinted:primaryCategory ?primary_category;
        vinted:secondaryCategory ?secondary_category;
        schema:brand ?item_brand.


} LIMIT 1
"""
for row in g.query(query_a):
    print(f"Name: {row.item_name}\nCategory: {row.primary_category}\nSubCategory: {row.secondary_category}\nBrand: {row.item_brand}")
```

```python
# Provide items from different subcategories that have the same brand.
query_b = """
PREFIX schema: <https://schema.org/> 
PREFIX vinted: <http://semanticweb2025.group5/vinted/>

SELECT ?brand_name ?item_name_1 ?item_subcategory_1 ?item_name_2 ?item_subcategory_2
WHERE {

    ?brand schema:name ?brand_name.

    ?item schema:name ?item_name_1;
        vinted:primaryCategory ?item_category_1;
        vinted:secondaryCategory ?item_subcategory_1;
        schema:brand ?brand.
        
    ?item2 schema:name ?item_name_2;
        vinted:primaryCategory ?item_category_2;
        vinted:secondaryCategory ?item_subcategory_2;
        schema:brand ?brand.


    FILTER(?item_subcategory_2!=?item_subcategory_1)

} 
"""

for row in g.query(query_b):
    print("\n")
    print(f"Brand: {row.brand_name}\nItem_1: {row.item_name_1}\nSubCategory_Item_1: {row.item_subcategory_1}\nItem_2: {row.item_name_2}\nSubCategory_Item_2: {row.item_subcategory_2}")
```
