```python
from rdflib import Graph, Literal, Namespace,URIRef
from rdflib.namespace import RDF, RDFS, XSD, SDO, OWL
import pandas as pd
import re
```

```python

g = Graph()
g.parse(fr"vinted_schema.ttl", format="turtle")
```

```python
print(g.serialize(format="turtle"))
```

```python
df = pd.read_csv("5_ken3140_webshop.csv")
df["Value 3"]=df["Value 3"].astype(int)
```

```python
SCHEMA = Namespace("https://schema.org/")
VINTED = Namespace("http://semanticweb2025.group5/vinted/")  

g.bind("schema", SCHEMA)
g.bind("ex", VINTED)
g.bind("rdfs", RDFS)
g.bind("xsd", XSD)
```

```python
def create_uri(input_string):
    """
    Creates a URI from an input string
    """
    cleaned_string = re.sub(r'[^\w]', '', input_string)
    uri_string = VINTED[cleaned_string]
    return uri_string
```

```python
# populate brands
unique_brands = list(filter(lambda x: type(x)==str,list(df["Value 1"].unique())))
unique_brands.append("BrandUnknown")
print(unique_brands)

for n in unique_brands:
    unique_brand_uri = create_uri(n)

    g.add((unique_brand_uri, RDF.type, SCHEMA.Brand))
    g.add((unique_brand_uri, SCHEMA.name, Literal(n, datatype=XSD.string)))

```

```python
for index, row in df.iterrows():

  category_name = row["Category"]
  sub_category_name = row["Subcategory"] 
  print(row["Item URI"][3:])
  item_name = row["Item Name"] #
  item_name_uri = create_uri(row["Item URI"][3:])
  print(item_name_uri)
  brand_uri = create_uri(row["Value 1"]) if type(row["Value 1"])==str else create_uri("BrandUnknown") #

  condition_name = row["Value 2"]
  price_name = float(row["Value 3"])
  color_name = row["Value 4"] if pd.notna(row["Value 4"]) and row["Value 4"] != "" else "not specified"
  size_name = row["Value 5"] if type(row["Value 5"])==str else "not specified"


  g.add((item_name_uri, RDF.type, SCHEMA.Product))
  g.add((item_name_uri, SCHEMA.name, Literal(item_name, datatype=XSD.string)))
  g.add((item_name_uri, SCHEMA.brand, brand_uri))

  g.add((item_name_uri, VINTED.primaryCategory, create_uri(category_name)))
  g.add((item_name_uri, VINTED.secondaryCategory, create_uri(sub_category_name)))

  g.add((item_name_uri, SCHEMA.itemCondition, Literal(condition_name, datatype=XSD.string)))
  g.add((item_name_uri, SCHEMA.price, Literal(price_name, datatype=XSD.decimal)))
  g.add((item_name_uri, SCHEMA.color, Literal(color_name, datatype=XSD.string)))
  g.add((item_name_uri, SCHEMA.size, Literal(size_name, datatype=XSD.string)))


print(g.serialize(format="ttl"))
```

```python
g.serialize(destination="5_ken3140_webshop.ttl", format="turtle")
print("Saved to 5_ken3140_webshop.ttl")
```
